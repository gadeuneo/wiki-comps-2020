```python
# Importing required packages/boilerplate code
import pandas as pd
import os
import sys
import time
import matplotlib.pyplot as pyplot
import numpy

from static_helpers import *

sys.path.append('../')

titles = get_titles()
titles_plus_talk = add_talk_pages(titles)
df_dictionary = {}
directory = "10 Year Revision Data"

file_names = [title + ".csv" for title in titles_plus_talk]

for file in file_names:
    complete_path = os.path.join(directory, file)
    
    if file_exists(complete_path):
        df_dictionary[file] = pd.read_csv(complete_path)
        
        # Adding another column to dataframe with coverted pythontime.
        df_dictionary[file]['pythontime'] = pd.to_datetime(df_dictionary[file]['timestamp'])
        df_dictionary[file].sort_values(by='pythontime', inplace=True)
        
```

```python
# Create a comprehensive data frame (comp_df)
all_df = []

# Making the dictionary into a list for concatenation
for key in df_dictionary.keys():
    all_df.append(df_dictionary[key])

all_df = pd.concat(all_df, ignore_index=True, sort=False)
all_df.sort_values(by='pythontime', inplace=True)

# Create a CSV file for the comp_df
all_df.to_csv("10 Year Updated.csv", encoding="utf-8")

all_df
```

```python
# Make data frame for anon and non-anon editors and print out the number of users
anon_comp_df = all_df[all_df['anon'] == True]
no_anon_comp_df = all_df[all_df['anon'] != True]
unique_editors = no_anon_comp_df['user'].nunique()
unique_anon_editors =anon_comp_df['user'].nunique()

print("Registered Editors: {0}, Anon Editors: {1}".format(unique_editors, unique_anon_editors))
```

```python
# Print out the average and median of the edit sizes
average_edit_size = all_df['size'].sum() // all_df['user'].count()

median_edit_size = int(all_df.median(axis=0)['size'])

print("Average Edit Size: {0} bytes, Median Edit Size: {1} bytes".format(average_edit_size, median_edit_size))
```
